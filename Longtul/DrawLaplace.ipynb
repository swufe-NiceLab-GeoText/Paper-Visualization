```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

def plot_multivariate_normal(mean, cov):
    """
    绘制二维多变量正态分布的等高线图

    参数:
    mean: 一维数组，二维正态分布的均值向量。 
    cov: 二维数组，协方差矩阵 

    配色：
    viridis 配色（从概率密度最低值到最高值，颜色依次呈现「深紫 → 靛蓝 → 青蓝 → 亮绿 → 明黄」的平滑过渡）
    """

    # 创建二维多变量正态分布
    mv_normal = multivariate_normal(mean=mean, cov=cov)

    #网格范围
    grid_range=(-3, 3)
    #网格分辨率
    grid_size=100

    cmap='viridis'

    # 生成网格点
    x, y = np.meshgrid(np.linspace(grid_range[0], grid_range[1], grid_size),
                       np.linspace(grid_range[0], grid_range[1], grid_size))
    pos = np.dstack((x, y))

    # 计算概率密度函数值
    z = mv_normal.pdf(pos)

    # 创建图形
    fig, ax = plt.subplots(figsize=(8, 6), facecolor='white')

    # 绘制等值面
    contour = ax.contourf(x, y, z, cmap=cmap, alpha=0.6, levels=np.linspace(0, np.max(z), 10))

    # 添加颜色条
    cbar = fig.colorbar(contour, ax=ax, shrink=0.5, aspect=5)
    cbar.ax.tick_params(labelsize=14)

    # 设置标题和标签
    ax.set_title('Multivariate Normal Distribution', fontsize=16)
    ax.set_xlabel('X', fontsize=14)
    ax.set_ylabel('Y', fontsize=14)

    # 去除坐标轴
    ax.axis('off')

    # 保存图像
    plt.savefig("multivariate_normal.png", transparent=True)

    plt.show()
   


```

```python
# 定义参数
mean = np.array([0, 0])  # 均值向量
cov = np.array([[1, 0.8], [0.8, 1]])  # 协方差矩阵
plot_multivariate_normal(mean,cov)
```
