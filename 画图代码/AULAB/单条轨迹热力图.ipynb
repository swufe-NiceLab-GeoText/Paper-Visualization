```python
import folium
from folium import plugins

def plot_1_pred(poi_data):
    """
    将 POI（兴趣点）数据可视化为交互式热力图

    参数：
    poi_data:嵌套列表，每个子列表包含两个字符串元素，分别表示 POI 的纬度和经度
    """
    # 创建基本地图
    m = folium.Map(zoom_start=13)

    # 将POI数据转换为热力图数据
    heat_data = [[float(row[0]), float(row[1]), 100] for row in poi_data]

    # 添加热力图层
    plugins.HeatMap(heat_data).add_to(m)

    # 显示地图
    m.save('1_pred.html')
    m
```
