```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_hyperparameter(data1, x_labels):
    """
    绘制多指标折线图，展示 3 个模型性能指标（Accuracy、Recall、F1）随某一超参数取值变化的趋势。
    
    参数：
    data1：二维数组，形状为 (3, n)，其中 n 为超参数取值数量，第 0 行对应Recall指标数据、第 1 行对应F1指标数据、第 2 行对应Accuracy指标数据。
    x_labels：列表，长度为 n（与 data1 的列数一致），每个元素为超参数的具体取值，作为 x 轴刻度标签。

    配色：sns.set_palette("pastel")
    """
    # 设置更加学术的配色方案
    sns.set_palette("pastel")
    
    # 创建图形和主坐标轴
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # 绘制折线图
    ax.plot(x_labels, data1[2], label='Accuracy', marker='o', color='#4e79a7', linestyle='-', linewidth=3, markersize=12)
    ax.plot(x_labels, data1[0], label='Recall', marker='s', color='#f28e2b', linestyle='--', linewidth=3, markersize=12)
    ax.plot(x_labels, data1[1], label='F1', marker='^', color='#e15759', linestyle=':', linewidth=3, markersize=12)
    
    # 设置x轴和y轴标签
    # ax.set_xlabel('(b) sample dimension', fontsize=30,fontname="Times New Roman")
    ax.set_ylabel('Metrics', fontsize=30)
    
    # 设置x轴刻度和标签
    plt.xticks(fontsize=30)
    
    # 设置图例
    ax.legend(loc='upper center', bbox_to_anchor=(0.8, 0.55), ncol=1, fontsize=24)
    
    # 调整刻度标签大小
    ax.tick_params(axis='both', which='major', labelsize=30)
    
    # 添加网格线
    ax.grid(True, linestyle='--', alpha=0.3)
    
    # 自适应设置y轴范围
    def set_axis_range(ax, data, padding=0.4):
        min_val = np.min(data)
        max_val = np.max(data)
        data_range = max_val - min_val
        ax.set_ylim(min_val - padding * data_range, max_val + padding * data_range)
    
    set_axis_range(ax, np.concatenate((data1[0], data1[1], data1[2])), padding=0.2)
    
    # 调整布局
    plt.tight_layout()
    
    # # 调整图表位置，使其更靠近中间
    # plt.subplots_adjust(top=0.85, bottom=0.15, left=0.1, right=0.9)
    
    # 保存图像
    plt.savefig('osm_d.pdf', format='pdf', dpi=300, bbox_inches='tight')
    plt.show()
```
