```python
import matplotlib.pyplot as plt
import numpy as np

def plot_TransitionMatrix(x,acc,recall,precision,mae,rmse):
    """
    绘制折线图，用于展示 X 轴变量对模型 5 类性能指标的影响趋势

    参数：
    x:数值列表,作为两个子图的 X 轴刻度
    acc:数值列表，每个元素对应 x 中对应位置的准确率数值
    recall：数值列表，每个元素对应 x 中对应位置的召回率数值
    precision：数值列表，每个元素对应 x 中对应位置的精确率（Precision） 数值
    mae：每个元素对应 x 中对应位置的平均绝对误差（MAE） 数值
    rmse：每个元素对应 x 中对应位置的均方根误差（RMSE） 数值

    配色：
    color='black' color='gray' color='blue' color='red' color='green'
    """

    # 创建画布和子图
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # 绘制第一个折线图（acc, recall, precision）
    ax1.plot(x, acc, marker='o', label='Accuracy', linewidth=4, markersize=15, color='black')
    ax1.plot(x, recall, marker='s', label='Recall', linewidth=4, markersize=15, color='gray')
    ax1.plot(x, precision, marker='D', label='Precision', linewidth=4, markersize=15, color='blue')
    
    ax1.legend(fontsize=18)
    ax1.set_xticks(x)  # 设置x轴刻度
    ax1.set_xticklabels(['0%', '20%', '40%', '60%', '80%'])  # 设置x轴刻度标签
    ax1.tick_params(axis='both', labelsize=20)  # 设置刻度字体大小
    
    # 自适应y轴
    ax1.set_ylim([min(min(acc), min(recall), min(precision)) - 0.04, max(max(acc), max(recall), max(precision)) + 0.04])
    
    # # 添加垂直网格
    # ax1.grid(axis='x', linewidth=1)
    
    # 绘制第二个折线图（mae, rmse）并设置双y轴
    ax2_mae = ax2  # 主y轴（左侧）
    ax2_rmse = ax2.twinx()  # 次y轴（右侧）
    
    # 绘制MAE（左侧y轴）
    ax2_mae.plot(x, mae, marker='o', label='MAE', linewidth=4, markersize=15, color='red')
    
    ax2_mae.tick_params(axis='y', labelcolor='red', labelsize=20)
    ax2_mae.set_xticks(x)  # 设置x轴刻度
    ax2_mae.set_xticklabels(['0%', '20%', '40%', '60%', '80%'])  # 设置x轴刻度标签
    ax2_mae.tick_params(axis='x', labelsize=20)  # 设置x轴刻度字体大小
    
    # 绘制RMSE（右侧y轴）
    ax2_rmse.plot(x, rmse, marker='s', label='RMSE', linewidth=4, markersize=15, color='green')
    ax2_rmse.tick_params(axis='y', labelcolor='green', labelsize=20)
    
    # 自适应y轴范围
    ax2_mae.set_ylim([min(mae) - 40, max(mae) + 40])  # MAE的y轴范围
    ax2_rmse.set_ylim([min(rmse) - 40, max(rmse) + 40])  # RMSE的y轴范围
    
    # 合并图例
    lines_mae, labels_mae = ax2_mae.get_legend_handles_labels()
    lines_rmse, labels_rmse = ax2_rmse.get_legend_handles_labels()
    ax2_mae.legend(lines_mae + lines_rmse, labels_mae + labels_rmse, fontsize=18)
    
    # # 添加垂直网格
    # ax2_mae.grid(axis='x', linewidth=1)
    
    # 显示图形
    plt.tight_layout()
    plt.savefig('折线.pdf', format='pdf', bbox_inches='tight')
    plt.show()
```

```python
# 示例数据
x = [0, 20, 40, 60, 80]
acc = [0.744630285852712, 0.7387502417995494, 0.7350521647046869, 0.7268961251628672, 0.7150592322846044]
recall = [0.7812930316586126, 0.7792202817210737, 0.7750849284389094, 0.7708857858929533, 0.7602986895993191]
precision = [0.8312635801005694, 0.8235289758171553, 0.8190174407773555, 0.8076598322118705, 0.7894692556066693]
mae = [234.2732010884587, 236.55083075999633, 237.63663667335996, 243.6382305382465, 252.44124418960936]
rmse = [777.8466439742521, 785.6507019631695, 788.4041139501986, 803.4464729862077, 811.0158828245613]

plot_TransitionMatrix(x,acc,recall,precision,mae,rmse)
```
